```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="my_app")

locations = ["madurai", "Reddiapatti"]
coordinates = []

for location in locations:
    try:
        loc = geolocator.geocode(location)
        if loc:
            coordinates.append((loc.latitude, loc.longitude))
        else:
            coordinates.append((None, None)) # Handle cases where location is not found
    except:
        coordinates.append((None, None)) # Handle potential errors during geocoding

print(coordinates)
```
